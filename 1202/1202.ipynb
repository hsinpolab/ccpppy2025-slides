# $\LaTeX$ 101

Sometimes you see `$2^2 + b^2 = c^2$` in my Jupyter notebook.

This is $\KaTeX$, the Jupyter way to write math.

The root of all evil is $\LaTeX$, which is an extension of $\TeX$.

Their relation is like C and C++, where La is the ++ part.

## Hello $\LaTeX$

```python
%%bash
cat > hello.tex << 'LATEXCODE'

\documentclass{article}
\begin{document}
    Hello, \LaTeX!
\end{document}

LATEXCODE
pdflatex hello.tex
```

## 嗨 XeLaTeX

Note that we use `pdflatex` to compile.

There are actually many compilers.

- `tex`: the original compiler developed by Knuth.
- `latex`: the extension of tex developed by Lamport.
- `pdftex`: the compiler that output pdf files.  (`tex` output dvi files)
- `pdflatex`: the compiler that output pdf files and with the la part.
- `xetex` and `xelatex`: the new compilers with unicode support; good for writing chinese; they are smart enough to output pdf.

```python
%%bash
cat > 嗨.tex << 'LATEXCODE'

\documentclass{article}
\usepackage{fontspec}
\setmainfont{Noto Sans TC}
\begin{document}
    嗨, Xe\LaTeX!
\end{document}

LATEXCODE
xelatex 嗨.tex
```

```python
%%bash
cat > magic.tex << 'LATEXCODE'
% !TeX encoding = UTF-8
% !TeX program = xelatex
\documentclass{article}
\usepackage{fontspec}
\setmainfont{Noto Sans TC}
\begin{document}
    magic, Xe\LaTeX!
\end{document}

LATEXCODE
xetex magic.tex
```

If you use `% !TeX program = ...` to specify the engine,
most IDE will respect it.

## Compiler Options

### -interaction=nonstopmode

Do not stop for errors.
Default behavior is to stop and wait for user's fix.
(This used to be useful when compiling time was long.)

### -file-line-error

Show error with file name and line number.

### -shell-escape

Enable the execution of external programs.
That is, allow `\write18{...}` to run commands in the OS shell.
This is disabled by default for security reasons.

### -synctex=1

Enable synchronization between the source and the output.

Command-left click in the PDF viewer will jump to the source line.

Command-option-J in the source file will jump to the output position.

In latex, fonts have three dimensions.

- The family dimension: `\rmfamily`, `\sffamily`, `\ttfamily`, or more.
- The weight dimension: bold `\bfseries`, not bold `\mdseries`, or in between.
- The shape dimension: Italic `\itshape`, slant `\slshape`, small capital `\scshape`, upright (default) `\upshape`

If you change shape, family and weight remain the same.

If you change weight, family and shape remain the same.

If you change family, shape and weight remain the same.

All these comes with `\textxx` version that applies changes locally.

(Non-text version = change from now on)

(text version = change inside the braces)

- `\textrm{rm}, \textsf{sf}, \texttt{tt}`,
- `\textbf{bold}`,
- `\textit{italic}, \textsl{slant}, \textsc{small capital}`.

Semantics: If you want to emphasize something, use `\emph{important!}`.

`\emph` is usually the same as `\textit` (italic).

But the semantic is to tell latex that this is important,
not just how you want to show this is important.

What if you are quoting Shakespeare and the character is speaking loudly?

In that case, you want to emphasize inside a block that is already emphasized.

`\emph` is smart enough to switch back to upright font.

That is the tradition of the typesetting industry:
if the environment is already speaking loudly,
then speaking gently is the way to pop out.

```python
%%bash
cat > emph.tex << 'LATEXCODE'

\documentclass{article}
\usepackage{url}
\begin{document}

    \url{https://www.youtube.com/watch?v=23hZQiMoB7A}

    Tim Minchin: \emph{To be \emph{OR} not to be.}

    Benedict Cumberbatch: \emph{To be or \emph{NOT} to be, that is the question.}

    David Tennant: \emph{To be or not to be, \emph{THAT} is the question.}

    Rory Kinnear: \emph{To be or not to be, \emph{IS} the question.}

    Sir Ian McKellen: \emph{To be or not to be, that is \emph{THE} question.}

    Dame Judi Dench: \emph{To be or not \emph{TO} be, that is the question.}

    Prince Charles (then Prince of Wales):
    \emph{To be or not to be, that is the \emph{QUESTION}.}

\end{document}

LATEXCODE
pdflatex emph.tex
```

Similar things happen to size control.

LaTeX likes semantics.

It doesn't want you to say 20pt font.

It wants you to say *larger* or *smaller*.

Each size level is 1.2x away from the precious one.

The compiler will control the precise size for you.

```python
%%bash
cat > size.tex << 'LATEXCODE'

\documentclass{article}
\begin{document}

    \tiny tiny
    
    \small small

    \normalsize normalsize

    \large large

    \Large Large

    \LARGE LARGE

    \huge huge

    \Huge Huge

\end{document}

LATEXCODE
pdflatex size.tex
```

## Dimensions

- 1pt is 1/72.27in (this is definition, not approximation)
- 1in is 2.54cm (this is definition, not approximation)
- 1cm is 0.01m (this is definition, not approximation)
- m is an SI unit; see wiki

- fill is a flexible unit, it will expand if there are space.
- if multiple fills are filling the same space, they share evenly.

- filll is a flexible unit that is more greedy.
- a positive amount of filll will always overpower any fill

- fil is a flexible unit that is less greedy than fill and filll.
- The end of every paragraph is a fil.

```python
%%bash
cat > length.tex << 'LATEXCODE'

\documentclass{article}
\usepackage{showframe}
\begin{document}

    []\hskip1cm[]\hskip2cm[]\hskip3cm[]\hskip4cm[]

    []\hskip0cm plus1fill[center]\hskip0cm plus1fill[]

    []\hskip0cm plus1fill[1:2 off center]\hskip0cm plus2fill[]

    []\hskip0cm plus1fill[]\hskip0cm plus2fill[]\hskip0cm plus3fill[]\hskip0cm plus4fill[]
    
    []\hskip0cm plus0.001filll[right]\hskip0cm plus1000fill[]

    []\hskip0cm plus1fill[center]%\hskip0cm plus1fill[]

\end{document}

LATEXCODE
pdflatex length.tex
```

A paragraph is a bunch of texts separated by empty lines.

The beginning of each paragraph is indented by `\parindent`.
Change this with `\parindent=2em`

The space between two paragraphs is controlled by `\parskip`.
Change this value with `\parskip=1em`

The spaces between two lines is controlled by
`\baselineskip` and `\lineskip`.

If your lines are shorter than `\baselineskip`,
LaTeX will stretch the space between two baselines to `\baselineskip`.
If not, LaTeX will make sure that the gap between two lines is `\lineskip`.

`em` is the width of the M character, the current value of font size.

`ex` is the height of the x character, about `0.5em`

## Document Structure

Documents have a tree-like structures 

- `\part{Part Title}` (only for books)
- `\chapter{Chapter Title}` (only for books)
- `\section{Section Title}` (the largest structure for journal and conference)
- `\subsection{Subsection Title}` (still useful)
- `\subsubsection{Subsubsection Title}` (not common)
- `\paragraph{Paragraph Title}` (handy if you three paragraphs, each explaining one thing.)
- `\subparagraph{Subparagraph Title}` (not common)

```python
%%bash
cat > structure.tex << 'LATEXCODE'

\documentclass{article}
\usepackage{lipsum}
\begin{document}

\section{Section One}
    \label{sec:intro}
    \lipsum[1]
\subsection{Section One Point One (Do Not Say Subsection One Point One)}
    \lipsum[2]
\subsubsection{Section One Point One Point One
    (Do Not Say Subsubsection One Point One Point One)}
    \label{sec:detail}
    \lipsum[3]
\subsection{Section One Point Two}
    \lipsum[4]
\section{Section Two}
    \label{sec:review}
    \lipsum[5]
\subsection{Section Two Point One}
    \lipsum[6]
\subsection{Section Two Point Two}
    \lipsum[7]

    Recall that Section~\ref{sec:intro} introduced the topic.
    Section~\ref{sec:detail} gave more details.
    Section~\ref{sec:review} reviewed past work.

\end{document}

LATEXCODE
pdflatex structure.tex
```

Use `\label` to remember the number of the last section/subsection/subsubsection/paragraph/subparagraph

## Inline Math

Math mixing with text is called *inline*.

```python
%%bash
cat > inline.tex << 'LATEXCODE'

\documentclass{article}
\usepackage{mathtools} % mathtools includes amsmath
\def\atan{\operatorname{atan}}
\begin{document}

    $\sin(\theta)$ is correct, $sin(theta)$ is not.

    $\atan(1/2)$ is correct, $\operatorname{atan}(1/2)$ is also correct.

    Around binary operator are 4mu spaces:
    $a + b$ is just ${a}\:{+}\:{b}$, which is like ${a}\mkern4mu{+}\mkern4mu{b}$.

    Around binary relation are 5mu space:
    $a = b$ is just ${a}\;{=}\;{b}$, which is like ${a}\mkern5mu{=}\mkern5mu{b}$.

\end{document}

LATEXCODE
pdflatex inline.tex
```

## Display Math

Math on its own line is called *display*.

```python
%%bash
cat > display.tex << 'LATEXCODE'

\documentclass{article}
\usepackage{mathtools} % mathtools includes amsmath
\begin{document}

    Use \verb|\[ ... \]|.
    Do not use \verb|$$ ... $$|, it is too old.
    \[ a^2 + b^2 = c^2. \]
    Put punctuation in the display math too.

    The \verb|equation*| environment from amsmath package
    is the same as \verb|\[ ... \]|.
    \begin{equation*}
        e^{i\pi} + 1 = 0.                                         \tag{Euler}
    \end{equation*}
    Use \verb|\tag{...}| to name or explain the equation.

    The \verb|align*| environment from amsmath package
    is for multiple equations that allow you to align verbs.
    (Each equal sign is a verb.)
    \begin{align*}
        f(x) &= (x + 1)^2 \\
            &= x^2 + 2x + 1.                                     \tag{Expand}
    \end{align*}
    The ampersand \verb|&| is the alignment point.
    The punctuation is needed for the last line.

    Use \verb|gather*| to center multiple equations without alignment.
    \begin{gather*}
        V - E + F = 2,                                         \tag{Euler} \\
        E = mc^2.                                              \tag{Einstein}
    \end{gather*}
    note that each line has a punctuation.

\end{document}

LATEXCODE
pdflatex display.tex
```

```python
%%bash
cat > label.tex << 'LATEXCODE'

\documentclass{article}
\usepackage{mathtools} % mathtools includes amsmath
\begin{document}

    The un-starred version \verb|equation|
    add numerical tags automatically.
    \begin{equation}
        e^{i\pi} + 1 = 0.                                       \label{Euler}
    \end{equation}
    Even if you don't add a label, the equation is still numbered.
    \begin{align}
        f(x) &= (x + 1)^2 \\
            &= x^2 + 2x + 1.                                  \label{Expand}
    \end{align}
    So you use \verb|\notag| to disable numbering on specific lines.
    \begin{gather}
        V - E + F = 2,                                              \notag \\
        E = mc^2.                                            \label{Einstein}
    \end{gather}
    note that each line has a punctuation.

    \eqref{Euler} is Euler's identity.

    \eqref{Expand} is the quadratic version of binomial theorem.

    \eqref{Einstein} is Einstein's mass-energy equivalence.

    Compile twice to get numbers right.

\end{document}

LATEXCODE
pdflatex label.tex
```

```python
%%bash
cat > bracket.tex << 'LATEXCODE'

\documentclass{article}
\usepackage{mathtools} % mathtools includes amsmath
\begin{document}
    Avoid using \verb|\left( ... \right)|
    \[
        \left\{ \left[ \left(
        \sum_{a,b\in A^2} p(a, b) \ln \frac{1}{p(a|b)}
        \right) \right] \right\}
    \]
    Instead, use \verb|\bigl, \Bigl, \biggl, \Biggl | to control the size
    \[
        \Biggl\{ \biggl[ \Bigl(
        \sum_{a,b\in A^2} p(a, b) \ln \frac{1}{p(a|b)}
        \Bigr) \biggr] \Biggr\}
    \]
\end{document}

LATEXCODE
pdflatex bracket.tex
```

## Slides

```python
%%bash
cat > beamer.tex << 'LATEXCODE'

\documentclass{beamer}
\begin{document}
\begin{frame}
    \frametitle{Poincaré conjecture}
    \begin{itemize}
        \item Every simply connected, closed 3-manifold is homeomorphic to the 3-sphere.
        \item Proven by Grigori Perelman in 2003.
        \item Uses Richard S. Hamilton's Ricci flow.
    \end{itemize}
\end{frame}

\begin{frame}
    \frametitle{Birch and Swinnerton-Dyer conjecture}

    \begin{itemize}
        \item <1-> Predicts a relationship between the rank of the group
            of rational points and the behavior of the L-function at \(s = 1\).
        \item <2-> Iwasawa theory with Euler systems (Kolyvagin--Rubin)
            is currently the most promising approach.
        \item <3-> Profound impact on number theory and cryptography;
            would clarify ranks of elliptic curves
            and reshape algorithms and security assumptions.
    \end{itemize}
\end{frame}

\begin{frame}
    \frametitle{Hodge conjecture}

    \begin{itemize}
        \item <+-> States that certain de Rham cohomology classes
            are algebraic.
        \item <+-> Motivic cohomology and derived algebraic geometry
            (Voevodsky--Deligne program) offer potential pathways.
        \item <+-> Would unify topology and algebraic geometry;
            enable algorithmic methods for classifying algebraic varieties
            and deepen understanding of the structure of complex manifolds.
    \end{itemize}
\end{frame}

\begin{frame}
    \frametitle{Navier–Stokes existence and smoothness}

    \begin{itemize} [<+->]
        \item Governs the motion of fluid substances.
        \item The existence and smoothness of solutions
            in three dimensions is unknown.
        \item Most promising method: convex integration for
            constructing wild weak solutions (De Lellis--Székelyhidi), and
            scale-critical regularity via harmonic analysis (Koch--Tataru).
    \end{itemize}
\end{frame}

\begin{frame}
    \frametitle{P versus NP}

    \begin{itemize} [<+- | alert@+>]
        \item Asks whether every efficiently verifiable problem
            is also efficiently solvable.
        \item Foundational in complexity theory;
            impacts optimization, cryptography, and AI.
        \item Key lines of attack: circuit lower bounds,
            barriers from natural proofs (Razborov--Rudich),
            and algebraic/geometric complexity (VP vs VNP).
        \item Consequences: if P $=$ NP, many cryptosystems collapse;
            if P $\neq$ NP, formal hardness underpins
            secure cryptography and explains intractability.
    \end{itemize}
\end{frame}
\end{document}

LATEXCODE
pdflatex beamer.tex
```

```python
%%bash
cat > foreach.tex << 'LATEXCODE'

\documentclass{article}
\usepackage{tikz}
\begin{document}
    \begin{tikzpicture}
        \foreach \x in {1, ..., 10} {
            \foreach \y in {1, ..., 25} {
                \pgfmathsetmacro\u{\x + \x/10}
                \pgfmathsetmacro\v{\y + (1 + sin(\y * 50)) / 2}
                \draw (\x, \y) rectangle (\u, \v);
            }
        }
    \end{tikzpicture}
\end{document}

LATEXCODE
pdflatex foreach.tex
```

```python
%%bash
cat > bubble.tex << 'LATEXCODE'

\documentclass{article}
\usepackage{tikz}
\begin{document}
    \begin{tikzpicture}
        \pgfmathsetseed{20251202}
        \foreach \x in {1, ..., 10} {
            \foreach \y in {1, ..., 15} {
                \pgfmathsetmacro\r{rnd} % random number in [0, 1]
                \pgfmathsetmacro\u{rnd/2} % random number in [0, 1/2]
                \pgfmathsetmacro\v{rand/3} % random number in [-1/3, 1/3]
                \pgfmathsetmacro\c{rnd*100} % random number in [0, 100]
                \draw [red!\c!blue] (\x + \u, \y + \v)
                    circle [x radius=\r, y radius=1-\r];
            }
        }
    \end{tikzpicture}
\end{document}

LATEXCODE
pdflatex bubble.tex
```

```python
%%bash
cat > EM.tex << 'LATEXCODE'

\documentclass[tikz]{standalone}
    \usetikzlibrary{3d} 
\begin{document}
\foreach \t in {0, ..., 35} {
    \pgfmathsetmacro\T{\t * 10 + 5}
    \pgfmathsetmacro\S{\T - 90}
    \pgfmathsetmacro\yv{cos(30)}
    \pgfmathsetmacro\zu{cos(\T)}
    \pgfmathsetmacro\zv{sin(30) * sin(\T)}
    \pgfmathsetmacro\xu{cos(\S)}
    \pgfmathsetmacro\xv{sin(30) * sin(\S)}
    \pgfmathsetmacro\A{cos(\t * 30 + 10)}
    \begin{tikzpicture} [x={(\xu cm, \xv cm)}, z={(\zu cm, \zv cm)}, y={(0 cm, \yv cm)}]
        \path (-10cm, -5cm) (10cm, 5cm);
        \tikzset{overlay}
        \def\wave{
            \draw[fill,thick,fill opacity=.2]
                plot [domain=-6:6,samples=500] ({\x}, {\A * sin(\x * 60)});
            \foreach \x in {-6, ..., 6} {
                \draw[thick] ({\x}, 0) -- ({\x}, {\A * sin(\x * 60)});
            }
        }
        \begin{scope}[canvas is zy plane at x=0,fill=blue]
            \wave
            \node at (6,-1.5) [transform shape] {magnetic field};
        \end{scope}
        \begin{scope}[canvas is zx plane at y=0,fill=red]
            \wave
            \node at (6,1.5) [rotate=180,xscale=-1,transform shape]
                {electric field};
        \end{scope}
    \end{tikzpicture}
}
\end{document}

LATEXCODE
pdflatex EM.tex
```
