```sage
%%bash
cat > add.cpp << CPPCODE

#include <iostream>
using namespace std;

int main() {
    int a = 3, b = 2;
    cout << a + b;
    cout << a + + b;
    cout << a ++ + b;
    cout << a + ++ b;
    cout << a ++ + + b;
    cout << a + + ++ b;
    cout << a ++ + ++ b;
}

CPPCODE
g++-14 -std=c++26 -Wall -Wextra -Wpedantic -Wshadow -Wconversion add.cpp -o add
./add
```

```sage
%%bash
cat > add.cpp << CPPCODE

#include <iostream>
using namespace std;

int main() {
    int a = 1, b = 3, c = 2;
    int *p = &a, *q = &b, *r = &c;
    int **x = &p, **y = &q, **z = &r;
    cout << **x << **y << **z << endl;
    swap(x, y);
    cout << **x << **y << **z << endl;
    swap(a, c); swap(q, r);
    cout << **x << **y << **z << endl;
    swap(*x, *z);
    cout << **x << **y << **y << endl;
    swap(*p, *q); swap(**y, **z);
    cout << **x << **y << **z << endl;
    cout << 1*a**p***x << endl;
}

CPPCODE
g++-14 -std=c++26 -Wall -Wextra -Wpedantic -Wshadow -Wconversion add.cpp -o add
./add
```

```sage
%%bash
cat > fun.cpp << CPPCODE

#include <iostream>
using namespace std;

float i(float x) { return 10 * x + 1; }
float i(int x) { return 10 * x + 2; }
int f(float x) { return 10 * x + 3; }
int f(int x) { return 10 * x + 4; }

int main() {
    cout << f(f(i(5))) << endl;
    cout << f(f(i(i(f(f(i(5.f))))))) << endl;
}

CPPCODE
g++-14 -std=c++26 -Wall -Wextra -Wpedantic -Wshadow -Wconversion fun.cpp -o fun
./fun
```

```sage
%%bash
cat > mt.cpp << CPPCODE

#include <iostream>
#include <random>
using namespace std;
int main () {
    mt19937 rand(20251022), rane(20251022), ranf(20251022);
    cout << rand() << " " << rand() << " " << rand() << endl;
    cout << rane() % 16 << endl;
    cout << ranf() % 16 << endl;
    cout << rane() * ranf() % 16 << endl;
    cout << ranf() * rane() % 16 << endl;
}

CPPCODE
g++-14 -std=c++26 -Wall -Wextra -Wpedantic -Wshadow -Wconversion mt.cpp -o mt
./mt
```

```sage
(r, c) = (10, 40)
table = [
    ['8' for _ in range(c)]
    for _ in range(r)
]

table[floor(random() * r)][floor(random() * c)] = '9'
table[floor(random() * r)][floor(random() * c)] = '9'
table[floor(random() * r)][floor(random() * c)] = '6'

for row in table:
    print(''.join(row))
```
