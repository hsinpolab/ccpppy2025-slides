## Olympic

- circles thick enough
- color roughly right (no need to use #123455).
- How circles overlap each other.

```sage
%%bash
cat > olympiad.tex << 'LATEXCODE'

\documentclass{article}
\usepackage{tikz}
\begin{document}
\begin{tikzpicture}
    \draw [line width=0.2cm, blue] (-2.5, 0) circle (1cm);
    \draw [line width=0.2cm, yellow] (-1.25, -1) circle (1cm);
    \draw [line width=0.2cm] (0, 0) circle (1cm);
    \draw [line width=0.2cm, green] (1.25, -1) circle (1cm);
    \draw [line width=0.2cm, red] (2.5, 0) circle (1cm);
    \begin{scope}
        \clip (-1.5, 0) circle (0.4);
        \draw [line width=0.2cm, blue] (-2.5, 0) circle (1cm);
    \end{scope}
    \begin{scope}
        \clip (-1, 0) circle (0.4);
        \draw [line width=0.2cm, yellow] (-1.25, -1) circle (1cm);
    \end{scope}
    \begin{scope}
        \clip (1, 0) circle (0.4);
        \draw [line width=0.2cm] (0, 0) circle (1cm);
    \end{scope}
    \begin{scope}
        \clip (1.5, 0) circle (0.4);
        \draw [line width=0.2cm, green] (1.25, -1) circle (1cm);
    \end{scope}
\end{tikzpicture}
\end{document}

LATEXCODE
pdflatex olympiad.tex
```

## Academic Sinica

- Institute name below
- DNA
- Atom

```sage
%%bash
cat > academia.tex << 'LATEXCODE'

\documentclass{article}
\usepackage{tikz}
\begin{document}
\begin{tikzpicture}
    \fill [blue!80!black] circle (5.5);
    \fill [yellow!50!gray!70!white] circle (4);
    \xdef\x{0}
    \foreach \A in {A, C, A, D, E, M, I, A} {
        \pgfmathsetmacro\x{\x + 1}
        \xdef\x{\x}
        \pgfmathsetmacro\t{\x * 10 - 45}
        \node [white, scale=2.5, rotate=\t] at ( -90 + \t: 4.75) {\A};
    }
    \foreach \s in {0, 120, 240} {
        \draw [line width=4pt, white]
            ellipse [x radius=3., y radius=0.5, rotate=\s];
    }
    \tikzset{shift={(-0.5, -3.2)}}
    \foreach \y in {1, 2, 3, 4} {
        \tikzset{rotate=45}
        \tikzset{shift={(0.7*\y, 0.7*\y)}}
        \fill [blue!30!white, rounded corners=5pt]
            (0.7, -0.7) -- (1, -0.7) -- (1, 1)
            (1, 1) -- (0.7, 1) -- (0.7, -0.7)
        ;
        \fill [blue!60!white, rounded corners=5pt]
            (0, 0) -- (1.7, 0) -- (1.7, 0.3)
            (1.7, 0.3) -- (0, 0.3) -- (0, 0)
        ;
    }
\end{tikzpicture}
\end{document}

LATEXCODE
pdflatex academia.tex
```

## ice
- nuclei
- electrons (Lewis structure)
- hydrogen bonds

```sage
%%bash
cat > ice.tex << 'LATEXCODE'

\documentclass{article}
\usepackage[a2paper, margin=1cm]{geometry}
\usepackage{tikz}
\begin{document}
\begin{tikzpicture}
    \foreach \x in {0, ..., 10} {
        \foreach \y in {0, ..., 10} {
            \ifodd\numexpr\x+\y\relax
            \tikzset{shift={(\x*3, \y*3)}}
            \fill [blue] (0, 0) circle (0.3) node [white, scale=0.3] {8n8p};
            \fill [magenta]
                (45:1.2) circle (0.1) node [white, scale=0.3] {p}
                [magenta] (135:1.2) circle (0.1) node [white, scale=0.3] {p}
            ;
            \draw [dotted] (0, 0) circle (0.6) circle (0.8);
            \draw [dotted] (45:1.2) circle (0.4) (135:1.2) circle (0.4);
            \fill [cyan]
                (45:0.8) +(-45:0.1) circle (0.05) +(135:0.1) circle (0.05)
                (135:0.8) +(45:0.1) circle (0.05) +(-135:0.1) circle (0.05)
                (-135:0.8) +(-45:0.1) circle (0.05) +(135:0.1) circle (0.05)
                (-45:0.8) +(45:0.1) circle (0.05) +(-135:0.1) circle (0.05)
            ;
            \draw [dash pattern=on2pt off1pt on2pt off3pt, <->, line width=2pt]
                (-45:1) -- (-45:2.5);
            \draw [dash pattern=on2pt off1pt on2pt off3pt, <->, line width=2pt]
                (-135:1) -- (-135:2.5);
            \fi
        }
    }
\end{tikzpicture}
\end{document}

LATEXCODE
pdflatex ice.tex
```

## Minecraft

- 3D perspective
- interesting terrain
- subtle color variations

```sage
%%bash
cat > minecraft.tex << 'LATEXCODE'

\documentclass[border = 1cm, tikz]{standalone}
	\usepackage{tikz-3dplot}
\begin{document}
	\makeatletter\let\oldpointxyz\pgfpointxyz
	\def\pgfpointxyz#1#2#3{
		\oldpointxyz{#1-\camerax}{#2-\cameray}{#3-\cameraz}
		\pgfmathsetmacro\depth{\rcarot*\pgftemp@x+\rcbrot*\pgftemp@y+\rccrot*\pgftemp@z}
		\pgfmathsetlength\pgf@x{\pgf@x*\cameras/(\camerad-\depth)}
		\pgfmathsetlength\pgf@y{\pgf@y*\cameras/(\camerad-\depth)}
	}\makeatother
	\tdplotsetmaincoords{50}{120}
	\def\camerax{0}\def\cameray{0}\def\cameraz{8}\def\camerad{40}\def\cameras{20}
	\begin{tikzpicture}[tdplot_main_coords]
		\def\block(#1,#2,#3)[#4,#5,#6];{
			\pgfmathsetmacro\a{#1}\pgfmathsetmacro\A{\a+1}
			\pgfmathsetmacro\b{#2}\pgfmathsetmacro\B{\b+1}
			\pgfmathsetmacro\c{#3}\pgfmathsetmacro\C{\c+1}
			\pgfmathsetmacro\rnd{rnd*10}
			\fill[.!\rnd!#4](\a,\b,\C)--(\A,\b,\C)--(\A,\B,\C)--(\a,\B,\C)--cycle;
			\fill[.!\rnd!#5](\A,\b,\c)--(\A,\B,\c)--(\A,\B,\C)--(\A,\b,\C)--cycle;
			\fill[.!\rnd!#6](\a,\B,\c)--(\a,\B,\C)--(\A,\B,\C)--(\A,\B,\c)--cycle;
		}
		\colorlet{g1}{green!80!.}\colorlet{d1}{brown!80!.}
		\colorlet{g2}{green!65!.}\colorlet{d2}{brown!65!.}
		\colorlet{g3}{green!50!.}\colorlet{d3}{brown!50!.}
		\def\grass(#1,#2,#3){\block(#1,#2,#3)[g1,g2,g3]}
		\def\dirt(#1,#2,#3){\block(#1,#2,#3)[d1,d2,d3]}
		\foreach \x in {-15, ..., 15}{
			\foreach \y in {-15, ..., 15}{
				\pgfmathtruncatemacro\radi{sqrt((\x)^2 + (\y)^2)}
				\ifdim\radi pt < 15pt
					\pgfmathtruncatemacro\Z{sin(\x*31-\y*19) * 3.5 +
					                        sin(\x*17+\y*26) * 2 + 5.5}
					\dirt(\x,\y,\Z);
					\dirt(\x,\y,\Z+1);
					\grass(\x,\y,\Z+2);
				\fi
			}
		}
	\end{tikzpicture}
\end{document}

LATEXCODE
pdflatex minecraft.tex
```
