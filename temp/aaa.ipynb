## Hilbert Curve

```
Alphabet : A, B
Constants : F + −
Axiom : A
Production rules:
A → +BF−AFA−FB+
B → −AF+BFB+FA−
```

````python
ll = 4
hstring = "A"
for it in range(ll):
    hstring = hstring.replace("A", "a")
    hstring = hstring.replace("B", "b")
    hstring = hstring.replace("a", "+BF-AFA-FB+")
    hstring = hstring.replace("b", "-AF+BFB+FA-")

hstring
````

````python
canvas = [[' ' for c in range(3*2^ll - 2)] for r in range(2*2^ll - 1)]
(x, y, dx, dy) = (0, 0, 0, 1)
for c in hstring:
    if c == 'A':
        canvas[x][y] = '+'
    if c == 'B':
        canvas[x][y] = '+'
    if c == 'F':
        x += dx
        y += dy
        if dx:
            canvas[x][y] = '|' 
        else:
            canvas[x][y] = '-' 
            x += dx
            y += dy
            canvas[x][y] = '-'
        x += dx
        y += dy
    elif c == '-':
        (dx, dy) = (-dy, dx)
    elif c == '+':
        (dx, dy) = (dy, -dx)
result = '\n'.join(''.join(row) for row in canvas)
print(result)
````

## Hanoi tower

````python
for ll in range(10):
    hstring = "A"
    for it in range(ll):
        hstring = hstring.replace("A", "a")
        hstring = hstring.replace("B", "b")
        hstring = hstring.replace("a", "-BF+A+FB-")
        hstring = hstring.replace("b", "+AF-B-FA+")
    canvas = [[' ' for c in range(4*2^ll - 3)] for r in range(2*2^ll - 1)]
    (x, y, dx, dy) = (0, 0, 2, 0)
    for c in hstring:
        if c == 'A':
            canvas[y][x] = '@'
        if c == 'B':
            canvas[y][x] = '@'
        if c == 'F':
            x = x + dx
            y = y + dy
            if dx * dy > 0: canvas[y][x] = '/'
            if dx * dy < 0: canvas[y][x] = '\\'
            if dx * dy == 0:
                canvas[y][x-1] = '-'
                canvas[y][x] = '-'
                canvas[y][x+1] = '-'
            x = x + dx
            y = y + dy
        elif c == '-':
            (dx, dy) = {
                (2, 0) : (1, 1),
                (1, 1) : (-1, 1),
                (-1, 1) : (-2, 0),
                (-2, 0) : (-1, -1),
                (-1, -1) : (1, -1),
                (1, -1) : (2, 0),
            }[dx, dy]
        elif c == '+':
            (dx, dy) = {
                (1, 1) : (2, 0),
                (-1, 1) : (1, 1),
                (-2, 0) : (-1, 1),
                (-1, -1) : (-2, 0),
                (1, -1) : (-1, -1),
                (2, 0) : (1, -1),
            }[dx, dy]
    canvas.reverse()
    result = '\n'.join(''.join(row) for row in canvas)
    with open(f'hanoi-{ll}.txt', 'w') as f:
        f.write(result)
````

````python
def decompose_word(w):
    if len(w) in [1, 2]:
        if w in elements:
            return [f'[{w}]']
        else: return []
    all_possibilities = []
    before1 = decompose_word(w[:1])
    after1 = decompose_word(w[1:])
    for a in before1:
        for b in after1:
            all_possibilities.append(a + b)
    before2 = decompose_word(w[:2])
    after2 = decompose_word(w[2:])
    for a in before2:
        for b in after2:
            all_possibilities.append(a + b)
    all_p = set(all_possibilities)
    all_p = list(all_p)
    all_p.sort()
    return all_p
````

## Benchmarking

````python
%%writefile fibench.cpp

#include <iostream>
#include <unistd.h>
#include <chrono>
using namespace std;

static volatile double garbage_can;
// C++ compiler will delete codes whose results are not used.
// This is a good thing for speeding up real programs.
// But for benchmarking, we want to perform the computation without using the results.
// So we use a "garbage can" to store the results in a way that
// the compiler cannot optimize the codes away.

template <class F>
double time_f_at_n (F&& f, long long n) {
    auto t0 = std::chrono::steady_clock::now();
    garbage_can /= 2;
    garbage_can += static_cast<double>(f(n)); // throw the result away
    auto t1 = std::chrono::steady_clock::now();
    long long ns = std::chrono::duration_cast<std::chrono::nanoseconds>(t1 - t0).count();
    return double(ns) / 1e9; // convert to seconds
}


// The recursion method
double fib_rec(long long n) {
    if (n <= 1) return double(n);
    return fib_rec(n-1) + fib_rec(n-2);
}

// Dynamic programming
double fib_dp(long long n) {
    if (n <= 1) return n;
    double  a = 0, b = 1, c = 0;
    for (long long i = 2; i <= n; ++i) {
        c = a + b; a = b; b = c;
    }
    return b;
}

// Matrix and square and multiply
struct M22 {
    double a, b, c, d;
};
M22 operator*(const M22& X, const M22& Y) {
    return {
        X.a*Y.a + X.b*Y.c,  X.a*Y.b + X.b*Y.d,
        X.c*Y.a + X.d*Y.c,  X.c*Y.b + X.d*Y.d
    };
};
double fib_matrix(long long n) {
    if (n == 0) return 0;
    M22 Q{1,1,1,0};
    M22 res{1,0,0,1}; // identity
    while (n) {
        if (n & 1) res = res * Q;
        Q = Q * Q;
        n >>= 1;
    }
    return res.b;
}
double fib_sleep(long long n) {
    usleep(n);
    return 1;
}

int main() {
    cout << "fib_rec spends " << time_f_at_n(fib_rec, 42) << " seconds" << endl;
    cout << "fib_dp spends " << time_f_at_n(fib_dp, 44442222) << " seconds" << endl;
    cout << "fib_matrix spends " << time_f_at_n(fib_matrix, 4444444422222222) << " seconds" << endl;
    cout << "fib_sleep spends " << time_f_at_n(fib_sleep, 5555555) << " seconds" << endl;
    cout << garbage_can << endl; // to avoid "unused variable" warning
}
````

````python
%%bash

g++-14 -std=c++26 -Wall -Wextra -Wpedantic fibench.cpp -o fibench
./fibench
````

````python
good_words = []

with open('words.txt', 'r') as f:
    for line in f:
        line = line.strip()
        line = line.lower()
        if set(line) <= set("abcdefghijklmnopqrstuvwxyz"):
            good_words.append(line)
# output good_words to good_words.txt
with open('good_words.txt', 'w') as f:
    for w in good_words:
        f.write(w + '\n')
````

````python
popularity = {}
for a in "abcdefghijklmnopqrstuvwxyz":
    popularity[a] = 0
    for b in "abcdefghijklmnopqrstuvwxyz":
        popularity[a + b] = 0
        for c in "abcdefghijklmnopqrstuvwxyz":
            popularity[a + b + c] = 0
            for d in "abcdefghijklmnopqrstuvwxyz":
                popularity[a + b + c + d] = 0
````

````python
word_count = 0

with open("good_words.txt") as f:
    for line in f:
        for j in range(len(line)):
            for i in range(max(0, j-4), j):
                fragment = line[i:j]
                popularity[fragment] += 1
        word_count += 1
        if word_count == int(sqrt(word_count)) ^ 2:
            print(word_count, sum(popularity.values()))
````

````python
pp = sorted([(ct, token) for token, ct in popularity.items() if ct > 0], reverse=True)
````
