### Binary Functions

```python
# Entry-wise operations

np.arange(10) * np.arange(10)
```

```python
# Multiplication table

A = np.arange(1, 11)
A0 = A[:, np.newaxis]
A0
```

```python
B = np.arange(1, 10)
B1 = B[np.newaxis, :]
B1
```

```python
A0 * B1
```

Explanation:
$$ A: [11] \to \mathbb N $$
$$ B: [10] \to \mathbb N $$
Then $A \cdot B$ is at best $[10] \to \mathbb N$.

But we make 
$$ A_0: [11] \times [10] \to \mathbb N $$
by
$$ A_0(x, y) = A(x). $$
Similarly, we make
$$ B_1: [11] \times [10] \to \mathbb N $$
by
$$ B_1(x, y) = B(y). $$
Then
$A_0 \cdot B_1: [11] \times [10] \to \mathbb N$ is what we want.

```python
# Multiplication is commutative

B1 * A0
```

```python
# Vandermonde matrix

(np.arange(7) - 3)[:, np.newaxis] ** np.arange(7)[np.newaxis, :]
```

```python
# Geometric sequence

np.exp(np.arange(10) * 2*np.pi*1j/10)
```

```python
# Two geometric sequences combined

np.exp(np.arange(10)[:, np.newaxis] * 2*np.pi*1j/10) ** np.arange(10)[np.newaxis, :]
```

This is the Fourier matrix
$$
\begin{bmatrix}
    1 & 1 & 1 & 1 & 1 & \ddots
    \\
    1 & \omega & \omega^2 & \omega^3 & \ddots
    \\
    1 & \omega^2 & \omega^4 & \ddots
    \\
    1 & \omega^3 & \ddots
    \\
    1 & \ddots
    \\
    \ddots
\end{bmatrix}
$$

### Broadcasting

```python
A = np.zeros([4, 5, 6, 7, 8, 9])
B = np.zeros([7, 8, 9])
A + B # They add?
```

```python
A = np.zeros([4, 5, 6, 7, 8, 9])
B = np.zeros([4, 5, 6])
A + B # Whaat happens?
```

If B.shape is a suffix of A.shape,
then B will be **broadcast** to A.shape.
That is,
$$ B'(a, b, c, x, y, z) = B(x, y, z). $$
It is also equivalent to 
`B_ = B[np.newaxis, np.newaxis, np.newaxis, :, :, :]`

Special case: If `B` is a scalar, `B.shape` is the empty tuple `()`,
which is a suffix of any shape. This means that a scalar can be broadcast to any shape.

That is to say, `A + 10` is a special case of broadcasting
where we define `ten[a, b, c] = 10` and compute `A + ten`.

But what if we really want to add `np.zeros([4, 5, 6])` to `np.zeros([4, 5, 6, 7, 8, 9])`?

```python
(A + B[:, :, :, np.newaxis, np.newaxis, np.newaxis]).shape
```

## einops

```python
import numpy as np
from einops import rearrange
```

```python
# Permuting the axis (rotating the hypercube)
# (may or may not change memory layout)

A = np.zeros([4, 5, 6, 7, 8, 9])
AA = rearrange(A, 'a b c x y z -> x y z a b c')
AA.shape
```

This is equivalent to letting
$$ A: [4] \times [5] \times [6] \times [7] \times [8] \times [9] \to \mathbb N $$
$$ A': [7] \times [8] \times [9] \times [4] \times [5] \times [6] \to \mathbb N $$
and $A'(x, y, z, a, b, c) = A(a, b, c, x, y, z)$.

```python
B = np.zeros([4, 5, 6])
(AA + B).shape

# Now they add because the suffixes of the shapes match
```

```python
# Arbitrary permutation works

rearrange(AA, 'a b c x y z -> a x b y c z').shape
```

```python
# 1 is equivalent to np.newaxis

BB = rearrange(B, 'a b c -> a b c 1 1 1')
BB.shape
```

```python
(A + BB).shape
```

Summary:
broadcasting will work if  
A.shape is a suffix of B.shape  
(or B.shape is a suffix of A.shape),  
and any `1` in shape is treated as a wildcard.

### repeat

```python
# More than one is repetition

from einops import repeat

repeat(np.arange(10), 'a -> a 3')
```

```python
# The repetition can happen in the same axis

repeat(np.arange(12).reshape([3, 4]), 'a b -> (a 5) b')
```

```python
# Reshape (flattening)

A = np.arange(24)
B = A.reshape([2, 3, 4])
C = rearrange(B, 'a b c -> (a b c)')
C == A
```

```python
# Unflattening

A = np.arange(24)
B = A.reshape([2, 3, 4])
D = rearrange(A, '(a b c) -> a b c', a=2, b=3, c=4)
(B == D).all()
```

```python
E = np.zeros([4, 5, 6])
rearrange(E, 'a b c -> (a b c)').shape
```

```python
rearrange(E, 'a b c -> (a b) c').shape
```

```python
rearrange(E, 'a b c -> a (b c)').shape
```

### pack/unpack

```python
# The generalization of np.stack
# And similar to np.block

from einops import pack, unpack

a = np.zeros([2, 3, 4])
b = np.zeros([2, 3, 5])
c = np.zeros([2, 3, 6])

packed, ps = pack([a, b, c], 'x y *')

packed.shape
```

```python
ps
```

```python
# Unpack is the reverse of pack

A, B, C = unpack(packed, ps, 'x y *')
A.shape, B.shape, C.shape
```

```python
a = np.zeros([2, 2, 11])
b = np.zeros([3, 3, 11])
c = np.zeros([4, 4, 11])

packed, ps = pack([a, b, c], '* z')

packed.shape
```

```python
ps
```

### einsum

```python
A = np.zeros([4, 5, 6])
B = np.zeros([6, 7])
np.einsum('wxy, yz -> wxz', A, B)
```

The sum is a function
$$ C: [4] \times [5] \times [7] \to \mathbb N $$
such that
$$ C(w, x, z) = \sum_{y=1}^6 A(w, x, y) \cdot B(y, z) $$

```python
reduce(rearrange(A, 'w x y -> w x y 1') * rearrange(B, 'y z -> 1 1 y z'), 'w x y z -> w x z', 'sum')
```

```python
# Trace
np.einsum('ii -> ', A)
```
