Use `//` to add comments

The compiler will ignore comments as if they were never there.
So the following program is the same as the previous one.
However comments are sentences that help you understand the code better.
A program that only machines can read is useless;
it also has to be readable by humans.

Therefore, make sure you add comments to your code
to explain what your code is doing.

````python
%%writefile integer_double.cpp

#include <iostream>
using namespace std;

int main() {
    int a = 1; // a is currently 1
    cout << a << endl;

    a = a + a; // a is currently 2
    cout << a << endl;

    a = a * 2; // a is currently 4
    cout << a << endl;
    
    a = 2 * a; // a is currently 8
    cout << a << endl;

    a += a; // a is currently 16
    cout << a << endl;

    a *= 2; // a is currently 32
    cout << a << endl;
    return 0;
}
````

````python
%%bash

g++-14 -std=c++26 integer_double.cpp -o id
./id
````

## How to take data from keyboard

`cin >>` is the opposite of `cout <<`.
The former takes input from the keyboard,
the latter outputs to the screen.

In the following code, we read one input from keyboard and store it in variable `a`.
We then print the value of `a` to the screen.
Afterward, we double `a` and print again.

````python
%%writefile integer_input.cpp

#include <iostream>
using namespace std;

int main() {
    int a = 1; // a is currently 1
    cout << "In the beginning, a is " << a << endl;
    
    cin >> a; // read a from keyboard, so a is not 1 anymore
    cout << "After cin, a is now " << a << endl;

    a /= 10;
    cout << "After division, a is now " << a << endl;

    int b = 2; // b is currently 2
    cout << "In the beginning, b is " << b << endl;

    cin >> b;
    cout << "After cin, b is now " << b << endl;

    b = a + b;
    cout << "After addition, b is now " << b << endl;

    return 0;
}
````

````python
%%bash

g++-14 -std=c++26 integer_input.cpp -o ii
./ii <<< "2025 303"
````

## A satisfactory game

Let's play a game.

Try to find 3 numbers $a$, $b$ and, $c$ such that
- $a$ is a prime
- $b$ is a power of $2$
- $c$ is a perfect square
- $b - a$ is even
- $c + 4a^2$ is also a square
- The last digit of $c$ is $a$


````python
%%writefile integer_game.cpp

#include <iostream>
using namespace std;

int main() {
    int a;
    int b;
    int c;
    cin >> a;
    cin >> b;
    cin >> c;
    
    cout << "Is a a prime? " << a << endl;
    cout << "Is b a power of 2? " << b << endl;
    cout << "Is c a perfect square? " << c << endl;
    cout << "Is b - a even? " << b - a << endl;
    cout << "Is c + 4a^2 also a square? " << c + 4 * a * a << endl;
    cout << "Is the last digit of c equal to a? " << c % 10 - a << endl;
    return 0;
}
````

````python
%%bash

g++-14 -std=c++26 integer_game.cpp -o ig
./ig <<< "12 13 14"
````

## Viewing Assembly Code

Earlier we learn that compilation is the process of
translating C++ source code to machine code.

While that is true, that is not the entire story.
There is a middle ground called assembly.
So the whole picture looks like

```text
C++ code --> Assembly --> Machine code
```

````python
%%bash

g++-14 -std=c++26 -S integer_double.cpp -o id.s
````

````python
%%bash

cat id.s
````

Why C is called C?
Because Assembly is A and there is another language B.
So the next letter available is C.
But B is no longer used to we only remember C and A.

Now, there is a very new thing called LLVM.
It's here

```text
C++ code --> LLVM --> Assembly --> Machine code
```

The fact that we need another layer of translation is very counterintuitive.
But there are two reasons.
(1) Other languages such as Rust and Julia can translate to LLVM

```text
Rust code --> LLVM --> Assembly --> Machine code
Julia code --> LLVM --> Assembly --> Machine code
```

Note that the part `LLVM --> Assembly --> Machine code` is shared,
so you only have to design the translator once.
The parts

```text
C++ code --> LLVM
Rust code --> LLVM
Julia code --> LLVM
```

need to be designed separately for each language,
but it is easier because LLVM is stronger than Assembly.

## Flow control

Use `if` and `else` to do different things depending on if a condition is met.

National ID rule

<http://www.mathland.idv.tw/life/nic.htm>

````python
%%writefile check_ID.cpp

#include <iostream>
#include <string>
using namespace std;

int main() {
    string id;
    cout << "Enter your National ID: ";
    cin >> id;

    // For the first character
    // A is ASCII 65, but should be 10 when checking
    // B is ASCII 66, but should be 11 when checking
    // C is ASCII 67, but should be 12 when checking
    // etc

    int letter = (id[0] % 32) + 9;
    int letter10 = letter / 10;
    int letter1 = letter % 10;

    // For the numbers
    // 0 is ASCII 48, but should be 0 when checking
    // 1 is ASCII 49, but should be 1 when checking
    // 2 is ASCII 50, but should be 2 when checking
    // etc

    int gender = id[1] - 48;
    int num1 = id[2] - 48;
    int num2 = id[3] - 48;
    int num3 = id[4] - 48;
    int num4 = id[5] - 48;
    int num5 = id[6] - 48;
    int num6 = id[7] - 48;
    int num7 = id[8] - 48;

    int check = id[9] - 48;

    int checksum =
        + 1 * letter10
        + 9 * letter1
        + 8 * gender
        + 7 * num1
        + 6 * num2
        + 5 * num3
        + 4 * num4
        + 3 * num5
        + 2 * num6
        + 1 * num7;
    
    if ((checksum + check) % 10 == 0) {
        cout << "correct national ID" << endl;
    }
    else {
        cout << "WRONG!" << endl;
    }

    return 0;
}
````

Bitwise operators

`|` bitwise-or (logical disjunction)  
`&` bitwise-and (logical conjunction)  
`^` bitwise-xor (exclusive-or) (mod-2 addition)  
`<<` shift left (multiply by 2)  
`>>` shift right (divide by 2)
