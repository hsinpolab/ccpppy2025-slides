## Binary representation

Any positive integer $n$ can be written as
$n = b_0 + 2b_1 + 4b_2 + 8b_3 + \dotsb + 2^kb_k$
where $b_i$ are the binary digits of $n$ and $k = \lfloor \log_2 n \rfloor$.

The operator `n >> s` shifts the binary representation of $n$
by $s$ bits to the right, discarding the $s$ least significant bits.
The operator `n << s` shifts the binary representation of $n$
by $s$ bits to the left, filling the $s$ least significant bits with zeros.

This is how you print the binary representation of an integer in C++

```C++
int n = 13;
cout << ((n >> 4) % 2)
cout << ((n >> 3) % 2)
cout << ((n >> 2) % 2)
cout << ((n >> 1) % 2)
cout << ((n >> 0) % 2)
```

Some examples include

1 = 1  
2 = 10  
3 = 11  
4 = 100  
5 = 101  
6 = 110  
7 = 111  

Other bitwise operators

`|` bitwise-or (logical disjunction)  
`&` bitwise-and (logical conjunction)  
`^` bitwise-xor (exclusive-or) (mod-2 addition)  
`<<` shift left (multiply by 2)  
`>>` shift right (divide by 2)  


## While loop

The following program keeps adding `1` to `n`

````sage
%%writefile while.cpp

#include <iostream>
#include <unistd.h>
using namespace std;

int main() {
    int n = 0;
    while (true) {
        n = n + 1;
        cout << n << endl;
        sleep(1);
    }
    return 0;
}
````

````sage
%%bash

g++-14 while.cpp -o w
./w
````

A while loop is useful because if can be stopped

````sage
%%writefile break.cpp

#include <iostream>
#include <unistd.h>
using namespace std;

int main() {
    int n = 0;
    while (true) {
        n = n + 1;
        cout << n << endl;
        sleep(1);

        if (n > 10) break;
    }
    return 0;
}
````

````sage
%%bash

g++-14 break.cpp -o b
./b
````

## Nested while loop

````sage
%%writefile square.cpp

#include <iostream>
#include <unistd.h>
using namespace std;

int main() {
    int m = 0;
    int n = 0;
    while (true) {
        m = m + 1;
        n = 0;
        while (true) {
            n = n + 1;
            cout << "+";
            usleep(100000);
            if (n > 10) break;
        }
        cout << endl;
        if (m > 10) break;
    }
    return 0;
}
````

````sage
%%bash

g++-14 square.cpp -o s
./s
````

## Estimate $\pi$

````sage
for x in range(-13, 13):
    for y in range(-13, 13):
        if x^2 + y^2 < 160: print('O', end=' ')
        else: print(' ', end=' ')
    print()
````
