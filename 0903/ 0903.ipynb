## Let's begin

`C/C++` is a compiled language.

You begin with some code that human can understand.

Here is an example: I create a file called `hello.cpp` insider a folder `0903`

````python
%%writefile hello.cpp

#include <iostream>
using namespace std;

int main() {
    cout << "Hello, World!" << endl;
    return 0;
}
````

Now I need to translate `hello.cpp` into something that machine can read.

In the following, I ask `g++-14` (the translator, usually called the **compiler**)
to translate `hello.cpp` into `hello` (the executable file).


````python
%%sh

g++-14 -std=c++26 hello.cpp -o hello
````

The executable program `hello` is very hard to read.
Here is its content

````python
%%sh

hexdump -C hello
````

Now les's execute that program.

In the following code,

* `./` stands for the current folder,
* `hello` is the name of the executable program we want to run.

````python
%%sh

./hello
````

## How to deal with numbers?

In the following code, `int` stands for integer,

`int a;` means that I want a new integer called `a`, and it is `20250903`.

Later, when I `cout << a`, I will get `20250903` on the screen.

````python
%%writefile integer_data.cpp

#include <iostream>
using namespace std;

int main() {
    int a = 20250903;
    cout << a;
    return 0;
}
````

````python
%%sh

g++-14 -std=c++26 integer_data.cpp -o id
./id
````

The program above is not fun because we already know it's going to be 20250903.

Let's make it more interesting by introducing some arithmetic.

````python
%%writefile integer_double.cpp

#include <iostream>
using namespace std;

int main() {
    int a = 1;
    cout << a << endl;
    a = a + a;
    cout << a << endl;
    a = a * 2;
    cout << a << endl;
    a = 2 * a;
    cout << a << endl;
    a += a;
    cout << a << endl;
    a *= 2;
    cout << a << endl;
    return 0;
}
````

````python
%%sh

g++-14 -std=c++26 integer_double.cpp -o id
./id
````

Use `//` to add comments

The compiler will ignore comments as if they were never there.
So the following program is the same as the previous one.
However comments are sentences that help you understand the code better.
A program that only machines can read is useless;
it also has to be readable by humans.

Therefore, make sure you add comments to your code
to explain what your code is doing.

````python
%%writefile integer_double.cpp

#include <iostream>
using namespace std;

int main() {
    int a = 1; // a is currently 1
    cout << a << endl;

    a = a + a; // a is currently 2
    cout << a << endl;

    a = a * 2; // a is currently 4
    cout << a << endl;
    
    a = 2 * a; // a is currently 8
    cout << a << endl;

    a += a; // a is currently 16
    cout << a << endl;

    a *= 2; // a is currently 32
    cout << a << endl;
    return 0;
}
````

````python
%%sh

g++-14 -std=c++26 integer_double.cpp -o id
./id
````

## How to take data from keyboard

`cin >>` is the opposite of `cout <<`.
The former takes input from the keyboard,
the latter outputs to the screen.

In the following code, we read one input from keyboard and store it in variable `a`.
We then print the value of `a` to the screen.
Afterward, we double `a` and print again.

````python
%%writefile integer_input.cpp

#include <iostream>
using namespace std;

int main() {
    int a = 1; // a is currently 1
    cout << "In the beginning, a is " << a << endl;
    
    cin >> a; // read a from keyboard, so a is not 1 anymore
    cout << "After cin, a is now " << a << endl;

    a /= 10;
    cout << "After division, a is now " << a << endl;

    int b = 2; // b is currently 2
    cout << "In the beginning, b is " << b << endl;

    cin >> b;
    cout << "After cin, b is now " << b << endl;

    b = a + b;
    cout << "After addition, b is now " << b << endl;

    return 0;
}
````

````python
%%sh

g++-14 -std=c++26 integer_input.cpp -o ii
./ii <<< "2025 303"
````

## A satisfactory game

Let's play a game.

Try to find 3 numbers $a$, $b$ and, $c$ such that
- $a$ is a prime
- $b$ is a power of $2$
- $c$ is a perfect square
- $b - a$ is even
- $c + 4a^2$ is also a square
- The last digit of $c$ is $a$


````python
%%writefile integer_game.cpp

#include <iostream>
using namespace std;

int main() {
    int a;
    int b;
    int c;
    cin >> a;
    cin >> b;
    cin >> c;
    
    cout << "Is a a prime? " << a << endl;
    cout << "Is b a power of 2? " << b << endl;
    cout << "Is c a perfect square? " << c << endl;
    cout << "Is b - a even? " << b - a << endl;
    cout << "Is c + 4a^2 also a square? " << c + 4 * a * a << endl;
    cout << "Is the last digit of c equal to a? " << c % 10 - a << endl;
    return 0;
}
````

````python
%%sh

g++-14 -std=c++26 integer_game.cpp -o ig
./ig <<< "12 13 14"
````

## File input and output

````python
%%writefile file_process.cpp

#include <iostream>
#include <fstream>
using namespace std;

int main() {
    ifstream fin("summands.txt");
    cout << "I am opening the file summands.txt" << endl;

    cout << "I am reading two integers" << endl;
    int a, b;
    fin >> a >> b;

    cout << "I am printing what is read" << endl;
    cout << a << " " << b << endl;

    cout << "I am trying to sum them" << endl;
    cout << a + b << endl;

    return 0;
}
````
